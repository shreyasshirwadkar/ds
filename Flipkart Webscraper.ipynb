# Converting the data to Text using soup

```python
soup = bs4.BeautifulSoup(request1.text)
soup
```

```python
reviews = soup.findAll('div', {'class': 'ZmyHeo'});
for review in reviews:
    print(review.get_text(), "\n")
```

```python
#average overall rating
ratings = soup.find('div',{'class':'ipqd2A'}).get_text();
print(ratings)
```

```python
#individual ratings

individual_ratings = soup.findAll('div',{'class':'XQDdHH Ga3i8K'});
for indi_rating in individual_ratings:
    print(indi_rating.get_text() + "\n");
```

```python
customer_name = soup.findAll('p',{'class':'_2NsDsF AwS1CA'});
for cust_name in customer_name:
    print(cust_name.get_text())
```

```python
data = []

# Loop through all items (assumes lengths are same or aligned)
for i in range(len(reviews)):
    try:
        review_text = reviews[i].text.strip()
        rating = individual_ratings[i].text.strip()
        name = customer_name[i].text.strip()

        data.append({
            "Customer Name": name,
            "Rating": rating,
            "Review": review_text
        })
    except IndexError:
        # In case of mismatch in lengths
        continue
```

```python
df = pd.DataFrame(data)
df.head()
```
