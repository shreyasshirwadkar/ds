```python
import pandas as pd
import matplotlib.pyplot as plt

# Histogram
df = pd.read_csv("adult_dataset.csv")
print(df.head())
plt.hist(df['age'],bins=20,color="skyblue",edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Dot Plot
import seaborn as sns
sns.stripplot(x='income',y='hours-per-week',data=df,jitter=True)
plt.title("Hours worked per week by Income class")
plt.show()
```

```python
# Bar Plot
df['education'].value_counts().plot(kind='bar',color='orange')
plt.title('Education Level Count')
plt.xlabel('Education')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Line Chart
edu_age = df.groupby('educational-num')['age'].mean()
edu_age.plot(kind='line',marker='o')
plt.title("Average Age by Education Number")
plt.xlabel("Education Level(Numeric)")
plt.ylabel("Average Age")
plt.grid(True)
plt.show()
```

```python
# Histogram and Scatter plot in Box Plot
import seaborn as sns
sns.boxplot(x='income',y='hours-per-week',data=df)
sns.stripplot(x='income',y='hours-per-week',data=df,color='red',jitter=0.2,size=2.5)
plt.title("Boxplot + Scatter for Income vs Hours-per-week")
plt.show()
```
